# Task-2 : Predict the optimum number of cluster of Iris Dataset

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
```

```python
#importing iris dataset and converting into pandas dataframe
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
#importing KMean for cluster analsis

from sklearn.cluster import KMeans
x = iris_df.iloc[:, [0, 1, 2, 3]].values
wcss = []

#Calculating cluster value through iteration
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the results onto a line graph
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
```

We can see from above 'Elbow Method',the optimum cluster value is '3'.

```python
# Applying kmeans to the dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
# Visualising the clusters between sepal length and sepal width
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],
            s = 70, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],
            s = 70, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 70, c = 'green', label = 'Iris-virginica');

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 70, c = 'yellow', label = 'Centroids');
plt.legend();
plt.figure(figsize = (12,8));
```

```python
# Visualising the clusters petal length and petal width
plt.scatter(x[y_kmeans == 0, 2], x[y_kmeans == 0, 3],
            s = 70, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 2], x[y_kmeans == 1, 3],
            s = 70, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 2], x[y_kmeans == 2, 3],
            s = 70, c = 'green', label = 'Iris-virginica');

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], 
            s = 70, c = 'yellow', label = 'Centroids');
plt.legend();
plt.figure(figsize = (12,8));
```
